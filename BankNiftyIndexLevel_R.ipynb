```python
def PriceRound(LTP,StepValue):
    return int(StepValue * round(float(LTP)/StepValue))
```

```python
# strike=strike.set_index('Symbol')
# #strike=strike.set_index(str('Symbol'))
# BNifty=pd.concat([BNifty,strike],join='inner',axis=1)

BNifty['Rounded Price']=BNifty.apply(lambda x: PriceRound(x['lastPrice'],x['Applicable Step value']),axis=1)
def LowerBand(LTP,RoundedPrice,StepValue):
    if LTP > RoundedPrice:
        return RoundedPrice
    else:
        return RoundedPrice-StepValue
BNifty['Lower Band']=BNifty.apply(lambda x: LowerBand(float(x['lastPrice']),x['Rounded Price'],x['Applicable Step value']),axis=1)
def UpperBand(LB,StepValue):
    return LB+StepValue
BNifty['Upper Band']=BNifty.apply(lambda x: UpperBand(int(x['Lower Band']),int(x['Applicable Step value'])),axis=1)
BNifty['Band%']=BNifty.apply(lambda x:x['Applicable Step value']/x['Rounded Price'],axis=1)
BNifty['Dist from LB']=BNifty.apply(lambda x: float(x['lastPrice'])-x['Lower Band'],axis=1)
BNifty['Dist from UB']=BNifty.apply(lambda x: x['Upper Band']-float(x['lastPrice']),axis=1)
BNifty['Dist from LB%']=BNifty.apply(lambda x:(-x ['Dist from LB'])/float(x['lastPrice']),axis=1)
BNifty['Dist from UB%']=BNifty.apply(lambda x:x ['Dist from UB']/float(x['lastPrice']),axis=1)
BNifty['WorstCase']=BNifty.apply(lambda x:x['Current']*(1+float(x['Dist from LB%'])),axis=1)
BNifty['BestCase']=BNifty.apply(lambda x:x['Current']*(1+float(x['Dist from UB%'])),axis=1)

''' note the objective of the  below function is to ensure that if the stock is near the lower bound it would get taken
    else if the stock is near the UB it gets taken '''


def Likely(DistLB,DistUB,CurIdx):
    if  -DistLB<DistUB: # -ve sign added to distLB to make comparison pluasible,else LB would always be lower than UB
        return (1+DistLB)*CurIdx
    else:
        return (1+DistUB)*CurIdx
BNifty['Likely']=BNifty.apply(lambda x:Likely(x['Dist from LB%'],x['Dist from UB%'],x['Current']),axis=1)

# to count the number of scripts close to UB/LB
def countLikely(WorstCase,BestCase,Likely):
    if WorstCase == Likely:
        return "Near LB"
    else:
        return "Near UB"
BNifty['LikelyCount'] = BNifty.apply(lambda x:countLikely(x['WorstCase'],x['BestCase'],x['Likely']),axis=1)
print (BNifty[['PreviousDay','Current','WorstCase','BestCase','Likely']].sum().round(2))

print(BNifty.LikelyCount.value_counts())
```

```python
%%R

suppressMessages(library(readxl))
suppressMessages(library(plyr)) # rounding based on strike price only *** round_any() *** function seem to work in R
# the below file contains
Bindx_lst=read_csv("/content/drive/MyDrive/Index_inputs/ind_niftybanklist.csv")

strike=read_excel('/content/drive/MyDrive/Index_inputs/sos_scheme.xls')
Bindx_order=data.frame(bankstk=c('AUBANK','AXISBANK','BANDHANBNK','BANKBARODA','FEDERALBNK','HDFCBANK','ICICIBANK','IDFCFIRSTB','INDUSINDBK','KOTAKBANK','PNB','SBIN'))#,'RBLBANK'

Bank_nifty_in<-left_join(Bindx_lst,bk_nifty,by=c("Symbol" = "symbol"))
Bank_nifty<-left_join(Bank_nifty_in,strike,by=c("Symbol" = "Symbol")) %>% dplyr::rename(StepValue = "Applicable Step value")

op_BkIdx<-as.numeric(df_bk$lastPrice[1]) # getting current price

bk_likely<-Bank_nifty %>% mutate(RoundedPrice= plyr::round_any(lastPrice,StepValue)) %>% rowwise() %>% 
  mutate(OpWeightedVal=op_BkIdx*Weightage/100) %>% 
  mutate(Lowerband= ifelse(lastPrice > RoundedPrice,RoundedPrice,RoundedPrice-StepValue)) %>% 
  mutate(UpperBand = Lowerband+StepValue) %>% 
  mutate('Band%' = StepValue/RoundedPrice) %>% 
  mutate('Dist from LB' = lastPrice-Lowerband) %>% 
  mutate('Dist from UB' = UpperBand-lastPrice) %>% 
  mutate('Dist from LB%' = `Dist from LB`/lastPrice) %>% 
  mutate('Dist from UB%' = `Dist from UB`/lastPrice) %>% 
  mutate(WorstCase= OpWeightedVal*(1-`Dist from LB%`)) %>% 
  mutate(BestCase = OpWeightedVal*(1+`Dist from UB%`)) %>% 
  mutate(Likely= ifelse((-`Dist from LB%`<`Dist from UB%`),OpWeightedVal*(1+`Dist from LB%`),OpWeightedVal*(1+`Dist from UB%`))) %>% 
  mutate(NextPrice=ifelse(`Dist from LB`<`Dist from UB`,Lowerband,UpperBand)) %>% 
  mutate(Pts_impact_index =op_BkIdx*(1+(Weightage/100*((NextPrice-lastPrice)/lastPrice)))) %>% 
# from Sheldon Natenberg methodology page 450
  mutate(Delta_pts=Pts_impact_index-op_BkIdx)

bn_pred<-data.frame(Probable =sum(bk_likely$Likely),Optimistic=sum(bk_likely$BestCase),Pessimistic=sum(bk_likely$WorstCase))


```
